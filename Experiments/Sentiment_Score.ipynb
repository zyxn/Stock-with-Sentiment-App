```python
import pandas as pd
import re
import ast
```

```python
df = pd.read_csv("C:\\TA\\Stock-with-Sentiment-App\\Dataset\\News\\News_Kompas.csv")
```

```python
df.info()
```

```python
# Fungsi untuk menghitung p(hi) dan n(hi) dengan mempertimbangkan huruf besar/kecil
def calculate_p_n(extract):
    if pd.isnull(extract):  # Jika nilai NaN
        return 0, 0
    try:
        # Ubah string menjadi dictionary
        extract_dict = ast.literal_eval(extract)
        sentiment = extract_dict.get('sentimen_berita', None)
        # Normalisasi nilai 'sentimen_berita' menjadi huruf kecil
        sentiment = str(sentiment).lower() if sentiment else None
        # Hitung p(hi) dan n(hi)
        p = 1 if sentiment == 'positif' else 0
        n = 1 if sentiment == 'negatif' else 0
        return p, n
    except (ValueError, SyntaxError):
        return 0, 0
```

```python
# Tambahkan kolom baru untuk p(hi) dan n(hi)
df[['p_hi', 'n_hi']] = df['Extract'].apply(lambda x: pd.Series(calculate_p_n(x)))

# Menampilkan contoh hasil
print(df[['Extract', 'p_hi', 'n_hi']].head())
```

```python
# Fungsi untuk mengekstrak tanggal
def extract_date(text):
    match = re.search(r'\d{2}/\d{2}/\d{4}', text)
    return match.group(0) if match else None
```

```python
# Ekstrak tanggal dari kolom dan ubah ke datetime
df['tanggal'] = df['tanggal'].apply(extract_date)
df['tanggal'] = pd.to_datetime(df['tanggal'], format='%d/%m/%Y', errors='coerce')
```

```python
df
```

```python
# Menghitung Daily Sentiment Score
daily_sentiment = df.groupby('tanggal').apply(
    lambda group: (group['p_hi'].sum() - group['n_hi'].sum()) / 
                  (group['p_hi'].sum() + group['n_hi'].sum()) 
                  
)
```

```python
df_sentiment_score = daily_sentiment.reset_index(name='Sentiment Score')
```
